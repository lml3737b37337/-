```python
import torch
# sigmoid
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

# sigmoid
def sigmoid_derivative(x):
    y = sigmoid(x)
    return y * (1 - x)

# x=0sigmoid
x = torch.tensor(0.0, requires_grad=True)  # 
sigmoid_val = sigmoid(x)
sigmoid_deriv = sigmoid_derivative(x)

# autograd
sigmoid_val.backward()  # sigmoid_valzz.grad

# 
print(f"Sigmoid at x=0: {sigmoid_val.item()}")
print(f"Derivative of Sigmoid at x=0 (manual calculation): {sigmoid_deriv.item()}")
print(f"Derivative of Sigmoid at x=0 (autograd): {x.grad.item()}")  # sigmoid_deriv.item()
```

```python
import torch

# 
x1 = torch.tensor(0.0, requires_grad=True)  # x1 = 0
x2 = torch.tensor(1.0, requires_grad=True)  # x2 = 1

# 
def model(x1, x2):
    term1 = torch.sigmoid(x1 + 2*x2 + 1)
    term2 = torch.sigmoid(x1 + (2*x2) + 1)
    y = torch.sigmoid(3 * term1 + term2 + 1)
    return y

# 
y = model(x1, x2)

# 
y.backward()

# 
print(f" x1  is: {x1.grad.item()}")
print(f" x2  is: {x2.grad.item()}")
```
